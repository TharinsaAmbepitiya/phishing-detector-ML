```python
import pandas as pd

# Convert date column to datetime format with UTC handling
df['date'] = pd.to_datetime(df['date'], errors='coerce', utc=True)
df['email_hour'] = df['date'].dt.hour


df[['date', 'email_hour']].head(10)
```

```python
# fill NaN values with empty strings to avoid errors
df['body'] = df['body'].fillna("").astype(str)
df['subject'] = df['subject'].fillna("").astype(str)

# Feature extraction for phishing email detection based on @Tharinsa's feature analysis report

# Extract the domain from sender email
df['sender_domain'] = df['sender'].apply(lambda x: x.split('@')[-1] if '@' in x else "")

df['email_length'] = df['body'].apply(len)  #number of characters in email body
df['subject_length'] = df['subject'].apply(len)  #number of characters in subject line
df['word_count'] = df['body'].apply(lambda x: len(x.split()))  #number of words in email body

# Special character analysis
df['num_special_chars'] = df['body'].apply(lambda x: sum(1 for char in x if not char.isalnum() and char != ' '))

# Count number of URLs in email body
df['num_urls'] = df['body'].apply(lambda x: x.count("http"))

# Flag emails with suspicious URLs
df['sus_url'] = df['body'].apply(lambda x: any(sub in x.lower() for sub in ['http://', 'bit.ly', 'tinyurl', 'paypa1.com']))

# Extract email timestamp anomaly
df['email_hour'] = pd.to_datetime(df['date'], errors='coerce').dt.hour  # Extract hour from timestamp


df.head(10)
```

**sender_domain** – Extracting the domain from the sender email helps identify suspicious domains, free email services, or misspelled company names. (e.g., loanofficertool.com vs. pobox.com).

**subject_length** – Phishing emails often use urgent or deceptive subject lines, which can be short and attention-grabbing or overly long and detailed.

**email_length**– Measures the total characters in the email body. Phishing emails may be concise with a single malicious link or long with misleading details.

**num_special_chars** – High usage of symbols (!, $, %, @) in phishing emails helps manipulate urgency and credibility.

**num_urls** – More URLs in an email increase phishing likelihood. Fraudulent emails often link to fake login pages.

**sus_url** – Checks if a URL contains risky patterns, such as HTTP instead of HTTPS, shortened URLs (bit.ly), or misspelled domains (paypa1.com).

**email_time_anomaly** – Phishing campaigns are often sent at odd hours to avoid detection (e.g., late at night or early morning bulk emails).

```python
df['receiver'].fillna("Unknown", inplace=True)
df['email_hour'].fillna(df['email_hour'].median(), inplace=True)
```

Since the number of unique values in the sender_domain column is really high tehre is no point of encoding, it is rather suitable to further extract feature based on sender_domain rather than using it straight away.

```python
df['domain_length'] = df['sender_domain'].apply(len)
df['domain_has_number'] = df['sender_domain'].apply(lambda x: any(char.isdigit() for char in x)).astype(int)

common_free_domains = ['gmail.com', 'yahoo.com', 'hotmail.com', 'outlook.com']
df['free_email_domain'] = df['sender_domain'].apply(lambda x: 1 if x in common_free_domains else 0)

df.head(10)
```

New domain-based numerical features:



*  domain_length (length of sender domain)
*  domain_has_number (binary flag: 1 if domain contains numbers, 0 otherwise)
*  free_email_domain (binary flag: 1 if free email provider, 0 otherwise)




```python
from sklearn.preprocessing import StandardScaler

# Define numerical features for scaling
num_features = ['email_length', 'subject_length', 'word_count', 'num_special_chars',
                'num_urls', 'email_hour', 'domain_length', 'domain_has_number', 'free_email_domain']

# Initialize scaler
scaler = StandardScaler()

# Apply normalization
df[num_features] = scaler.fit_transform(df[num_features])

df.head()
```

Normalization ensures that all numerical features have a similar scale.ML models perform better when features are standardized. The StandardScaler transformed all numerical values to have zero mean and unit variance.
